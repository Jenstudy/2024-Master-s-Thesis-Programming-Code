````python
from sdv.datasets.local import load_csvs
from sdv.datasets.demo import download_demo

# This is the default folder name that the GOogle Colab notebook uses.
# Change this if you have your own folder with CSV files.
FOLDER_NAME = 'content/'

try:
  datasets = load_csvs(folder_name='/content/')
except ValueError:
  print('You have not uploaded any csv files. Using some demo data instead.')
  datasets, _ = download_demo(
    modality='multi_table',
    dataset_name='fake_hotels'
  )
````

````python
real_data = datasets['R_DATA']
````

````python
real_data
````

````python
from sdv.metadata import SingleTableMetadata

metadata = SingleTableMetadata()
````

````python
metadata.detect_from_csv('R_DATA.csv')
````

````python
metadata.visualize()
````

````python
metadata.update_column(
    column_name='CHILD-NUM',
    sdtype='id')
    # regex_format='U_[0-9]{3}')
````

````python
metadata.update_column(
    column_name='YN',
    sdtype='boolean')
````

````python
metadata.update_column(
    column_name='MLU-W',
    sdtype='numerical',
    computer_representation='Float')
````

````python
metadata.update_column(
    column_name='MLU-C',
    sdtype='numerical',
    computer_representation='Float')
````

````python
metadata.update_column(
    column_name='MLU5-W',
    sdtype='numerical',
    computer_representation='Float')
````

````python
metadata.update_column(
    column_name='MLU5-C',
    sdtype='numerical',
    computer_representation='Float')
````

````python
metadata.update_column(
    column_name='VOCD-D',
    sdtype='numerical',
    computer_representation='Float')
````

````python
metadata.update_column(
    column_name='VOCD-D-C',
    sdtype='numerical',
    computer_representation='Float')
````

````python
metadata.update_column(
    column_name='CTTR-W',
    sdtype='numerical',
    computer_representation='Float')
````

````python
metadata.update_column(
    column_name='CTTR-C',
    sdtype='numerical',
    computer_representation='Float')
````

````python
metadata.set_primary_key(column_name='CHILD-NUM')
````

````python
metadata.visualize()
````

````python
real_data
````

````python
from sdv.single_table import CTGANSynthesizer
synthesizer = CTGANSynthesizer(
    metadata,
    enforce_rounding=False,
    epochs=15000,
    verbose=False
)
synthesizer.fit(real_data)
````

````python
synthesizer.save('ctgan_model_f15.sav')
````

````python
synthetic_data = synthesizer.sample(num_rows=4854)
````

````python
synthetic_data.head(4854)
````

````python
synthetic_data.to_csv("F15-4854")
````





```
評估
作為合成資料專案的最後一步，可以根據真實資料評估和視覺化合成資料
```


````python
from sdv.evaluation.single_table import evaluate_quality

quality_report = evaluate_quality(
    real_data,
    synthetic_data,
    metadata)
````


```
運行診斷
使用此函數可以接收有關合成資料的一些診斷結果。檢查合成欄位是否為真實資料的純副本、合成資料是否涵蓋整個值範圍、合成資料是否符合原始範圍。

```



````python
from sdv.evaluation.single_table import run_diagnostic

diagnostic_report = run_diagnostic(
    real_data=real_data,
    synthetic_data=synthetic_data,
    metadata=metadata)
````
